#BestBuy Corp

One of our group’s ideas is an electronics store corporation. We use BestBuy as an example of a familiar electronics store.  

The electronics store idea is a unique example in that it focuses on the corporation as a whole, as opposed to a singular BestBuy.

On a typical day at the BestBuy corporation headquarters, data is sent from multiple store locations and product suppliers. From the data that is collected, important decisions are made. 

1.  Customers

  a. Name - Letters
  
  b. Email - E-mail format
  
  c. Contact details

  d. Number of purchases - Number from 0 to 2^64

  e. Payment method - Selected from a list

  f. Store credit - Number from 0 to 2^32
	
The customers table allows the stores to hold onto some customer details for future reference. Being able to hold some of this information is important because if there is ever a problem with product the customer has purchased then the stores are able to know when it was purchased, what the product was, and how this product was purchased by the customer. Along with this, keeping the customers name and email address, along with contact details allows them to get a hold of them for any reason.

2. Workers

  a. Name - Letters

  b. Id - 9 digit number

  c. Seniority - Number from 0 to 80

  d. Salary - Number from 0 to 1M (not minimum wage)

  e. Store ID (store that he works at) - Number
	
The Workers table allows the company to keep track of all their employees easier. It keeps their name and worker id all together as well as some more corporate information such as their salary and seniority. It also shows which store they work at due to the store ID.

3. Suppliers

  a. Name - Letters
  
  b. Phone number - 3 digits in parenthesis, followed by 7 digits
  
  c. Email - E-mail format
  
  d. Address
  
  e. Balance - Number
	
The Suppliers table allows the BestBuy corp to keep track of all their different suppliers. It not only tells them who the supplier is but also gives them all the information they would need to contact them, such as phone number and email. It also shows them the address of where the suppliers are located.

4. Products (end user)

  a. ID - 9 digit number
  
  b. Name - Letters
  
  c. Retail Price - Number
  
  d. Supplier - Letters
  
  e. Number of Units Sold - Number
	
The Products table allows BestBuy corp to inventory all of their products they sell in store and online. Each product has an ID that correlates with its name through BestBuy corp. Along with the ID, this table shows the retail price of each product and the supplier of the product. Lastly is shows how many units were sold to customers.

5. Stores

  a. ID - 9 digit number

  b. Address

  c. Number of workers - Number

  d. Manager ID - 9 digit number

  e. Store’s phone number - 3 digits in parenthesis, followed by 7 digits

The stores table allows BestBuy corp to ID all of their stores together as well as showing them the address of each store. It also gives the number of workers currently working at a certain store and the manager who is running the store. This table also gives the store’s phone number for any specific needs they may have.	

### BestBuy's Goal
BestBuy goal is, “to figure which suppliers are supplying the products that sell most units”.

BestBuy will use the tables “Products” and “Suppliers” to answer this question.
